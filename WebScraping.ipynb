```python
import requests
```

```python
result = requests.get("https://en.wikipedia.org/wiki/Jonas_Salk")
```

```python
result.text
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(result.text,'lxml')
```

```python
soup
```

```python
soup.select("title")
```

```python
soup.select("title")[0]
```

```python
soup.select("title")[0].getText()
```

```python
soup.select(".toctext")
```

```python
for item in soup.select(".toctext"):
    print(item.text)
```
